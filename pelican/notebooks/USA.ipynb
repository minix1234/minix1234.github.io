```python
import pandas as pd
import numpy as np
```

```python

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'  #time_series_19-covid-Confirmed.csv'#
dfcases = pd.read_csv(url,index_col=0,parse_dates=[0])

threshold = 250  #used to remove countries in earlier infections from plots
#dfcases = dfcases[dfcases['Country/Region'] == 'Canada']
dfcases = dfcases.reset_index()
dfcases = dfcases.drop(columns=['UID','iso2','iso3','code3','FIPS','Admin2',"Province_State",'Country_Region','Lat',"Long_"])
```

```python
#dfcases = pd.read_csv('time_series_19-covid-Confirmed.csv')
dfcases
```

```python
#df = dfcases.groupby('Province/State').sum()
#df = df.T
#df = df.reset_index()
#df = df.drop([0, 1])
#df['index'] = df['index'].astype('datetime64')
#df = df.set_index('index')


df = dfcases.rename(columns={'Province_State ': 'date'})

df = df.T
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df = df.reset_index()
df = df.rename(columns={'index': 'dates'})
df['dates'] = df['dates'].astype('datetime64')
df = df.set_index('dates')
df = df.apply(pd.to_numeric) # convert all columns of DataFrame
df = df[df.index >= '2020-03-01']
#df.head()
```

```python
dfmax = pd.DataFrame(df.max(),columns=['max'] )
dfmax = dfmax.reset_index()
dfmax = dfmax.rename(columns={'date': 'country'})
#dfmax = dfmax.set_index('country')
dfmax
```

```python
dfmax = dfmax[dfmax['max'] > threshold]
llist = dfmax['Combined_Key'].values.tolist()
df1 = df[llist]
df1.head()
```

```python
#df = df.reset_index()

#pd.to_timedelta(df.index - pd.to_datetime('2020-01-22'))

#df1.index = pd.to_timedelta(df.index - pd.to_datetime('2020-01-22')).days
#df1
```

```python
dfmelt = df1.reset_index()
dfmelt = dfmelt.melt(id_vars=['dates'])

dfmelt = dfmelt.rename(columns={'Combined_Key': 'City_State'})
dfmelt = dfmelt.rename(columns={'value': 'cases'})
#dfmelt = dfmelt.drop(columns=['dates'])
dfmelt = dfmelt.set_index('dates')
dfmelt['City_State'] = dfmelt['City_State'].astype('category')
#dfmelt['cases'] = dfmelt['cases'].astype('int32') # fails due to nan
dfmelt.dtypes
```

```python
import plotly.express as px
```

```python
fig = px.line(dfmelt, x=dfmelt.index, y="cases",color='City_State',log_y=True)
fig.update_layout(
    title="Aligned Covid-19 Active Cases Data",
    xaxis_title="Dates",
    yaxis_title="Number of Active Cases",
    font=dict(
        family="Courier New, monospace",
        size=16,
        color="#7f7f7f"
    ),
    legend=dict(
        font=dict(
            family="sans-serif",
            size=12,
            color="black"
        )
    )
)
fig.show()
fig.write_html("State.html")
```
